```python
#Importa bibliotecas necessárias

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

```python
'''
Data Acquisition
'''

#Busca as informações da origem dos dados
df = pd.read_csv('ArtistMusics.csv', sep=',')
df.head()
```

```python
'''
Data Preparation
'''

#Seleciona as colunas desejadas e transforma elas em um numpy array
f1 = df['TotalTracks'].values
f2 = df['TotalValue'].values
x = np.array(list(zip(f1,f2)))
```

```python
'''
Data exploration
'''

x
```

```python
kmeans = KMeans(n_clusters=3).fit(x)
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
distance = kmeans.fit_transform(x)
distance
```

```python
plt.scatter(x[:, 0], x[:,1], s = 100, c = kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'red',label = 'Centroids')
plt.show()
```

```python
df['cluster']  = kmeans.labels_
df
```
